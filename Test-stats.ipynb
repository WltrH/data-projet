# Tests sur les données JSON

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Lecture des données json TOP10
df = pd.read_json('json/top10.json')
df.head()
```

```python
#Prix actuels et variations récentes des prix des crypto-monnaies classées par capitalisation boursière
plt.figure(figsize=(20,10))
sns.barplot(x='market_cap', y='name', data=df, palette='rocket')
plt.title('Prix actuels et variations récentes des prix des crypto-monnaies classées par capitalisation boursière', fontsize=20)
plt.xlabel('Prix actuel en USD', fontsize=15)
plt.ylabel('Nom de la crypto-monnaie', fontsize=15)
plt.legend()
plt.show()
```

```python
#Capitalisation boursière totale des crypto-monnaies avec comparaison avec le marché des 10 crypto-monnaies les plus chères
plt.figure(figsize=(20,10))
sns.barplot(x='market_cap', y='name', data=df, palette='rocket')
plt.title('Capitalisation boursière totale des crypto-monnaies avec comparaison avec le marché des 10 crypto-monnaies les plus chères', fontsize=20)
plt.xlabel('Capitalisation boursière totale en USD', fontsize=15)
plt.ylabel('Nom de la crypto-monnaie', fontsize=15)
plt.legend()
plt.show()
```
